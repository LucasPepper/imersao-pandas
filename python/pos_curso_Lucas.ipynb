## Joins

* merge com left, right, inner, etc.
* 'on' como chave

* Falar de concat

```python
df_votacao.head(2)
```

```python
df_group.head(2)
```

```python
"""
O SQL abaixo é equivalente ao df_votacao.merge() da próxima célula

SELECT t1.*,
    t2.total_votos AS total_votos_partido
    
    FROM tb_votacao AS t1
    
    LEFT JOIN df_group AS t2
    ON t1.ano_eleicao = t2.ano_eleicao
    AND t1.numero_turno = t2.numero_turno
    AND t1.sigla_partido = t2.sigla_partido
    
"""
```

```python
df_group = df_group.reset_index()
df_vot_join = df_votacao.merge(right=df_group, 
                on=["ano_eleicao", "numero_turno", "sigla_partido"],
                how="left",
                suffixes=["", "_partido"])
```

```python
df_vot_join["representacao_partido"] = df_vot_join["total_votos"] / df_vot_join["total_votos_partido"]
df_vot_join
```
